```python
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt
import numpy as np
```

Создаем выборку размером 1000 с 5-ю степенями свободы, строим гистограмму и график плотности распределения.

```python
df = 5
r = chi2.rvs(df, size=1000)
```

```python
p = chi2.pdf(r, df, loc=0, scale=1)
```

```python
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
plt.plot(x, chi2.pdf(x, df), lw=5, alpha=0.6, label='chi2 pdf', color='#FFDD44')

plt.hist(r, density=True, histtype='stepfilled', alpha=0.2)
plt.legend(loc='best', frameon=False)
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

Сгенерируем 1000 выборок объёма 5 и построим гистограмму распределения их средних значений (mean)

```python
arr = []
for i in range(1000):
    arr.append(chi2.rvs(df, size=5).mean())
plt.hist(arr, density=True, histtype='stepfilled', alpha=0.2)
#вот тут не очень разобрался: оранжевым цветом - плотность распределения вероятностей исходного распределение,
#красным - сгенерированного
plt.plot(x, norm.pdf(x, df), lw=5, alpha=0.6, label='chi2 pdf', color='orange')
plt.plot(arr, norm.pdf(arr, df), lw=5, alpha=0.2, label='chi2 pdf', color='red')
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

Получившаяся гистограмма лишь отдаленно напоминает нормальное распределение (только потому что мы знаем цель задания), пока что скорее хи-квадрат или Бернулли.

Сгенерируем 1000 выборок объёма 10 и построим гистограмму распределения их средних значений (mean).

```python
arr = []
for i in range(1000):
    arr.append(chi2.rvs(df, size=10).mean())
plt.hist(arr, density=True, histtype='stepfilled', alpha=0.2)

#вот тут не очень разобрался: оранжевым цветом - плотность распределения вероятностей исходного распределение,
#красным - сгенерированного
plt.plot(x, norm.pdf(x, df), lw=5, alpha=0.6, label='chi2 pdf', color='orange')
plt.plot(arr, norm.pdf(arr, df), lw=5, alpha=0.2, label='chi2 pdf', color='red')

plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

Получившаяся гистограмма намного ближе к нормальному распределению.
Сгенерируем 1000 выборок объёма 50 и построим гистограмму распределения их средних значений (mean).

```python
arr = []
for i in range(1000):
    arr.append(chi2.rvs(df, size=50).mean())
plt.hist(arr, density=True, histtype='stepfilled', alpha=0.2)

#вот тут не очень разобрался: оранжевым цветом - плотность распределения вероятностей исходного распределение,
#красным - сгенерированного
plt.plot(x, norm.pdf(x, df), lw=5, alpha=0.6, label='chi2 pdf', color='orange')
plt.plot(arr, norm.pdf(arr, df), lw=5, alpha=0.2, label='chi2 pdf', color='red')

plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

Данный набор значений средних тысячи выборок по 50 элементов распределен нормально.

В распределении хи-квадрат дисперсия равна удвоенному количеству степеней свободы (определено как df)

```python
chi2.var(df, loc=0, scale=1)
```

В распределении хи-квадрат среднее значение равно значению матожидания и количеству степеней свободы

```python
chi2.mean(df, loc=0, scale=1)
```
